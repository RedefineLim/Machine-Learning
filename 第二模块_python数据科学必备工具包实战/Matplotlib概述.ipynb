```python
np.random.seed(0)
x = np.arange(5)
y = np.random.randint(-5,5,5)

fig,axes = plt.subplots(ncols=2)
v_bars = axes[0].bar(x,y,color='red')
h_bars = axes[1].barh(x,y,color='blue')

axes[0].axhline(0, color='white', linewidth=1)
axes[1].axvline(0, color='white', linewidth=1)

plt.show()
```

```python
fig,ax = plt.subplots()
v_bars = ax.bar(x,y,color='lightblue')
for bar,height in zip(v_bars, y):
    if height < 0:
        bar.set(edgecolor='darkred',color='green',linewidth=3)
plt.show()
```

```python
x = np.random.randn(100).cumsum()
y = np.linspace(0,10,100)

fig,ax = plt.subplots()
ax.fill_between(x,y,color='lightblue')
plt.show()
```

# 条形图

```python
mean_values = [1,2,3]
variance = [0.2,0.4,0.5]
bar_label = ['bar1', 'bar2', 'bar3']

plt.bar(x_pos, mean_values,yerr=variance, alpha=1)
x_pos = list(range(len(bar_label)))
max_y = max(zip(mean_values, variance))
plt.ylim([0,(max_y[0]+max_y[1])*1.2])
plt.ylabel('variable y')
plt.xticks(x_pos, bar_label)
plt.show()
```

```python
x1 = np.array([1,2,3])
x2 = np.array([2,2,3])

bar_labels = ['bar1','bar2','bar3']
fig = plt.figure(figsize=(8,6))
y_pos = np.arange(len(x1))
y_pos = [x for x in y_pos]

plt.barh(y_pos, x1, color='g', alpha=0.5)
plt.barh(y_pos, -x1, color='b', alpha=0.5)

plt.show()
```

```python
green_data = [1,2,3]
blue_data = [3,2,1]
red_data = [2,3,3]
labels = ['group1', 'group2', 'group3']

pos = list(range(len(green_data)))
width = 0.2
fig,ax = plt.subplots(figsize = (8,6))

plt.bar(pos,green_data, width, alpha=0.5, color='g', label=labels[0])
plt.bar([p+width for p in pos],blue_data, width, alpha=0.5, color='b', label=labels[1])
plt.bar([p+width*2 for p in pos],red_data, width, alpha=0.5, color='r', label=labels[2])
plt.show()
```

```python
data = range(200,225,5)
bar_labels = ['a','b','c','d','e']

fig = plt.figure(figsize=(10,8))

y_pos = np.arange(len(data))

plt.yticks(y_pos, bar_labels, fontsize=16)

bars = plt.barh(y_pos, data,alpha=0.5, color='g')

for b,d in zip(bars, data):
    plt.text(b.get_width() + b.get_width()*0.05, b.get_y() + b.get_height()/2, '{0:.2%}'.format(d/min(data)))

plt.show()
```

```python
data = range(200,225,5)
bar_labels = ['a','b','c','d','e']

fig = plt.figure(figsize=(10,8))

y_pos = np.arange(len(data))

plt.yticks(y_pos, bar_labels, fontsize=16)

bars = plt.barh(y_pos, data,alpha=0.5, color='g')

plt.vlines(min(data), -1, len(data)+0.5, linestyles='dashed',color='black')

for b,d in zip(bars, data):
    plt.text(b.get_width() + b.get_width()*0.05, b.get_y() + b.get_height()/2, '{0:.2%}'.format(d/min(data)),color='black')

plt.show()
```

```python
mean_values = range(10,18)
x_pos = range(len(mean_values))

import matplotlib.colors as col
import matplotlib.cm as cm

cmap1 = cm.ScalarMappable(col.Normalize(min(mean_values), max(mean_values), cm.hot))
cmap2 = cm.ScalarMappable(col.Normalize(0,20,cm.hot))

plt.subplot(121)
plt.bar(x_pos, mean_values, color=cmap1.to_rgba(mean_values))

plt.subplot(122)
plt.bar(x_pos, mean_values, color=cmap2.to_rgba(mean_values))

plt.show()
```

```python
patterns = ('-', '+', 'x', '\\', '*', 'o', '--', '.')

fig = plt.gca()

mean_value = range(1,len(patterns) + 1)
x_pos = list(range(len(mean_value)))

bars = plt.bar(x_pos, mean_value)

for bar,pattern in zip(bars,patterns):
    bar.set_hatch(pattern)
plt.show()
```

# 盒图

```python
guo_data = [np.random.normal(0, std, 100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
plt.boxplot(guo_data, notch=False, sym='s', vert=True)

plt.xticks([ y+1 for y in range(len(guo_data))], ['x1', 'x2', 'x3'])
plt.xlabel('x')
plt.title('box plot')
```

```python
guo_data = [np.random.normal(0, std, 100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
bplot = plt.boxplot(guo_data, notch=False, sym='s', vert=False)

plt.yticks([ y+1 for y in range(len(guo_data))], ['x1', 'x2', 'x3'])
plt.ylabel('x')
plt.title('box plot', color='black')

for components in bplot.keys():
    for line in bplot[components]:
        line.set_color('black')
```

```python
guo_data = [np.random.normal(0, std, 100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
bplot = plt.boxplot(guo_data, notch=True, sym='s', vert=True)

plt.yticks([ y+1 for y in range(len(guo_data))], ['x1', 'x2', 'x3'])
plt.ylabel('x')
plt.title('box plot', color='black')

for components in bplot.keys():
    for line in bplot[components]:
        line.set_color('black')
```

```python
guo_data = [np.random.normal(0, std, 100) for std in range(1,4)]
fig = plt.figure(figsize=(8,6))
bplot = plt.boxplot(guo_data, notch=True, sym='s', vert=True,patch_artist=True)

plt.yticks([ y+1 for y in range(len(guo_data))], ['x1', 'x2', 'x3'])
plt.ylabel('x')
plt.title('box plot', color='black')

for components in bplot.keys():
    for line in bplot[components]:
        line.set_color('black')
        
colors = ['pink', 'lightblue', 'lightgreen']
for pathch, color in zip(bplot['boxes'], colors):
    pathch.set_facecolor(color)
```

# 小提琴图violinplot

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
guo_data = [np.random.normal(0,std,100) for std in range(6,10)]

axes[0].violinplot(guo_data, showmeans=False, showmedians=True)
axes[0].set_title('violin plot', color='black')

axes[1].boxplot(guo_data)
axes[1].set_title('box plot', color='black')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,5))
guo_data = [np.random.normal(0,std,100) for std in range(6,10)]

axes[0].violinplot(guo_data, showmeans=False, showmedians=True)
axes[0].set_title('violin plot', color='black')

axes[1].boxplot(guo_data)
axes[1].set_title('box plot', color='black')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(guo_data))])
    
plt.setp(axes, xticks=[y+1 for y in range(len(guo_data))], xticklabels=['x1', 'x2', 'x3','x4'])
```

```python
import math

x = np.random.normal(loc=0.0, scale=1.0, size=300)
width =0.5
bins = np.arange(math.floor(x.min())-width, math.ceil(x.max())+width, width)

ax = plt.subplot(111)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tick_params(bottom='off', top='off', left='off')
plt.grid(True)

plt.hist(x, alpha=0.5, bins=bins)
```
