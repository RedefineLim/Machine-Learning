# 直方图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(0,20,1000)
```

```python
bins = np.arange(-100,100,5)

plt.hist(data, bins=bins, color='yellow')
plt.xlim([min(data)-5, max(data)+5])
plt.show()
```

```python
import random
data1 = [random.gauss(15,10) for i in range(500)]
data2 = [random.gauss(5,5) for i in range(500)]

bins = np.arange(-50,50,2.5)
plt.hist(data1, bins=bins,label='class1',alpha=0.3)
plt.hist(data2, bins=bins,label='class2',alpha=0.3)
plt.legend(loc='best')

plt.show()
```

```python
mu_vec1 = np.array([0,0])
cov_mat1 = np.array([[2,0],[0,2]])

x1_samples = np.random.multivariate_normal(mu_vec1, cov_mat1, 100)
x2_samples = np.random.multivariate_normal(mu_vec1+0.2, cov_mat1+0.2, 100)
x3_samples = np.random.multivariate_normal(mu_vec1+0.4, cov_mat1+0.4, 100)

plt.figure(figsize=(8,6))
plt.scatter(x1_samples[:,0], x1_samples[:,1], marker='x', color='blue', alpha=0.3,label='x1')
plt.scatter(x2_samples[:,0], x2_samples[:,1], marker='o', color='red', alpha=0.3,label='x1')
plt.scatter(x3_samples[:,0], x3_samples[:,1], marker='^', color='green', alpha=0.3,label='x1')

plt.legend(loc='upper left')
plt.show()
```

```python
x_coords = [0.13,0.22,0.36,0.56,0.68,0.74,0.93]
y_coords = [0.75,0.34,0.44,0.53,0.80,0.25,0.55]

plt.figure(figsize=(8,6))
plt.scatter(x_coords, y_coords, marker='s',s=50)

for x,y in zip(x_coords, y_coords):
    plt.annotate('(%s, %s)'%(x,y), xy=(x,y),xytext=(x,y+0.1), textcoords='offset points', ha='center')

plt.show()
```

```python
mu_vec1 = np.array([0,0])
cov_mat1 = np.array([[1,0],[0,1]])
X = np.random.multivariate_normal(mu_vec1, cov_mat1, 500)

fig = plt.figure(figsize=(8,6))

R = X**2
R_sum = R.sum(axis=1)
plt.scatter(X[:,0], X[:,1], color='grey',marker='o',s=20*R_sum,alpha=0.5)
plt.show()
```

# 3D图

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)

X,Y = np.meshgrid(x,y)

Z = np.sin(np.sqrt(X**2+Y**2))
ax.plot_surface(X,Y,Z,rstride=1, cstride=1, cmap='rainbow')
ax.contour(X,Y,Z, zdim='z', offset=-2, cmap='rainbow')

ax.set_zlim(-2,2)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

theta = np.linspace(-4*np.pi, 4*np.pi, 100)
z = np.linspace(-2,2,100)
r = z**2 + 1
x = r*np.sin(theta)
y = r*np.cos(theta)

ax.plot(x,y,z)
plt.show()
```

```python
np.random.seed(1)
def randrange(n,vmin,vmax):
    return (vmax-vmin)*np.random.rand(n)+vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
n = 100
for c,m,zlow,zhigh in [('r','o',-50,-25),('b','x',-30,-5)]:
    xs = randrange(n,23,32)
    ys = randrange(n,0,100)
    zs = randrange(n,zlow,zhigh)
    ax.scatter(xs,ys,zs, color=c, marker=m)
plt.show()
```

```python
m = 51251
f = 40742

m_perc = m/(m+f)
f_perc = f/(m+f)

colors = ['navy','lightcoral']
labels = ['Male','Female']

plt.figure(figsize=(8,8))
paches, texts, autotexts = plt.pie([m_perc,f_perc], labels=labels, autopct='%1.1f%%', explode=[0,0.1],colors=colors)

for text in texts + autotexts:
    text.set_fontsize(20)
for text in autotexts:
    text.set_color('white')
```

# Pandas与sklearn结合实例

```python
import pandas as pd
```

```python
np.random.seed(0)
df = pd.DataFrame({'Condition 1':np.random.rand(20),
                  'Condition 2':np.random.rand(20)*0.9,
                  'Condition 3':np.random.rand(20)*1.1})
df.head()
```

```python
fig,ax = plt.subplots()
df.plot.bar(ax=ax,stacked=False)
plt.show()
```

```python
fig,ax = plt.subplots()
df.plot.bar(ax=ax,stacked=True)
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
df_ratio = df.div(df.sum(axis=1),axis=0)
fig,ax = plt.subplots()

df_ratio.plot.bar(ax=ax, stacked=True)
ax.yaxis.set_major_formatter(FuncFormatter(lambda y,_ : '{:.0%}'.format(y)))

plt.show()
```
